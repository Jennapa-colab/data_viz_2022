```python
import numpy as np #นำเข้าข้อมูล
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
from math import pi #เอามาใช้เพื่อดึงค่าของพาย
```

```python
data = {'client_id' : ['c1' , 'c2' , 'c3' , 'c4'],
         'Math' : [10,5,4,7],
         'Eng' : [7,1,3,5],
         'Friendly' : [4,8,10,4],
        'Look' : [4,8,7,6],
       'Money' : [2,10,5,9]}
df = pd.DataFrame.from_dict(data)
```

```python
df
```

```python
categories = list(df.columns[1:])  
categories
#ดูประเภททั้งหมด และ ตัดแถว client_id 
```
