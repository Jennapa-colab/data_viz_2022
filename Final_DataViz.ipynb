# Data 2 hospital

```python
data2 = pd.read_excel(os.path.join(path,'hospital_list.xlsx'))
data2
```

```python
data_h2 = data2[['hcode','name','zone_name','province','amphur','tambon','name5','short_english']]      # เลือกเฉพาะคอลัมน์ที่ต้องการ
data_h2
```

```python
data_h2.isnull().sum() 
```

```python
data_h2['name'] = data_h2['name'].fillna('ไม่ระบุ') 
data_h2['zone_name'] = data_h2['zone_name'].fillna('ไม่ระบุ') 
data_h2['amphur'] = data_h2['amphur'].fillna('ไม่ระบุ') 
data_h2['tambon'] = data_h2['tambon'].fillna('ไม่ระบุ') 
data_h2['name5'] = data_h2['name5'].fillna('ไม่ระบุ') 
data_h2['short_english'] = data_h2['short_english'].fillna('ไม่ระบุ');
```

```python
data_h2.isnull().sum() 
```

```python
data_df = data_drop2.merge(data_h2,how='left',left_on='site',right_on='hcode')     # เชื่อมตาราง
data_df
```

```python
data_df.isnull().any()  # ตรวจสอบหาค่า missing
```

จากการตรวจสอบพบว่าข้อมูลไม่มีค่า missing แล้ว และข้อมูลพร้อมที่จะนำไปทำ dashboard
